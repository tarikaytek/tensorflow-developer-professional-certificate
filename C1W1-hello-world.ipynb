```python
import numpy as np
import tensorflow as tf 
from tensorflow import keras
```

```python
#creating model which has 1 layer with 1 neuron
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
#compiling model with optimizer stochastic gradient descent and loss function mean squared error
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# x to y mapping function y=3x-2
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype='float')
y = np.array([-8.0, -5.0, -2.0, 1.0, 4.0, 7.0, 10.0], dtype='float')
```

```python
#training
model.fit(x=x, y=y, epochs=1000)
```

```python
#prediction
model.predict([3.0, -4.0, 10])
```
